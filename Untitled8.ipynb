```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
OFFLOAD_FOLDER = "/content/falcon_offload"

TEACHER_MODEL = "tiiuae/Falcon-H1-3B-Instruct"

# Tokenizer
teacher_tokenizer = AutoTokenizer.from_pretrained(TEACHER_MODEL, use_fast=True)

# Model
teacher_model = AutoModelForCausalLM.from_pretrained(
    TEACHER_MODEL,
    device_map="auto",
    offload_folder=OFFLOAD_FOLDER,
    offload_state_dict=True,
    low_cpu_mem_usage=True,
    torch_dtype=torch.float16
)
```

```python
def build_prompt(passage):
    return f"""
You are a careful legal annotator.
Passage: {passage}

Task:
Decide whether the HIGHLIGHTED PASSAGE contains bias by a judicial actor (judge, arbitrator, prosecutor, juror) against a party or a class of persons.
Return a JSON object EXACTLY with these keys:
{{"label":"Y" or "N", "prob":float(0.0-1.0), "explanation":"one-sentence justification quoting the key phrase if present"}}
"""

def generate_label(passage):
    prompt = build_prompt(passage)
    inputs = teacher_tokenizer(prompt, return_tensors="pt").to(DEVICE)
    with torch.no_grad():
        outputs = teacher_model.generate(
            **inputs,
            max_new_tokens=100,
            do_sample=False
        )
    text_out = teacher_tokenizer.decode(outputs[0], skip_special_tokens=True)
    return text_out
```

```python
import json
from tqdm import tqdm

# Load dataset
with open("challenge_dataset.json", "r", encoding="utf-8") as f:
    data = json.load(f)

results = []

for item in tqdm(data):
    passage = item.get("textpassage", "")
    try:
        teacher_out = generate_label(passage)
    except Exception as e:
        teacher_out = '{"label":"N","prob":0.5,"explanation":"Fallback due to error."}'

    results.append({
        "instance_id": item.get("instance_id"),
        "textpassage": passage,
        "teacher_output": teacher_out
    })

# Save results
with open("teacher_labels_falcon.jsonl", "w", encoding="utf-8") as f:
    for rec in results:
        f.write(json.dumps(rec, ensure_ascii=False) + "\n")
```
